```python
import pandas as pd 

wilshire = pd.read_csv('wilshire.csv')
trn = wilshire[wilshire['date'] <= '2017-01-01']
tst = wilshire[wilshire['date'] >= '2017-01-01']
tst = tst.iloc[:, 1:]

tst
```

```python
from ecm import ecm
from ecmpredict import ecmpredict
```

```python
m1 = ecm(trn['Wilshire5000'], xeq=trn[['CorpProfits']], xtr=trn[['UnempRate']])
m1.summary()
```

```python
m2 = ecm(trn['Wilshire5000'], xeq=trn[['CorpProfits']], xtr=trn[['UnempRate']], includeIntercept=False)
m2.summary()
```

```python
m3 = ecm(trn['Wilshire5000'], xeq=None, xtr=trn[['CorpProfits']])
m3.summary()
```

```python
m4 = ecm(trn['Wilshire5000'], xeq=trn[['CorpProfits']], xtr=None, includeIntercept=False)
m4.summary()
```

```python
m5 = ecm(trn['Wilshire5000'], xeq=trn[['CorpProfits']], xtr=trn[['UnempRate']], weights=trn['FedFundsRate'])
m5.summary()
```

```python
ecmpredict(m1, tst, tst['Wilshire5000'].iloc[0])
```

```python
ecmpredict(m2, tst, tst['Wilshire5000'].iloc[0])
```

```python
ecmpredict(m3, tst, tst['Wilshire5000'].iloc[0])
```

```python
ecmpredict(m4, tst, tst['Wilshire5000'].iloc[0])
```

```python
ecmpredict(m5, tst, tst['Wilshire5000'].iloc[0])
```
